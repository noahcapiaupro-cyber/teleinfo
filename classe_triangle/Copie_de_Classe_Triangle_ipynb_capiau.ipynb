```python
import math

class Point:
    def __init__(self):
        self.x = 0
        self.y = 0

    def deplacer(self, deltaX, deltaY):
        self.x += deltaX
        self.y += deltaY

    def afficher(self):
        print("x =", self.x)
        print("y =", self.y)


class Segment:
    def __init__(self, A=None, B=None):
        self.A = A if A else Point()
        self.B = B if B else Point()

    def longueur(self):
        res = (self.A.x - self.B.x)**2 + (self.A.y - self.B.y)**2
        return math.sqrt(res)

    def milieu(self):
        M = Point()
        M.x = (self.B.x + self.A.x) / 2
        M.y = (self.B.y + self.A.y) / 2
        return M
```

### A - Coder une classe **Triangle** contenant 3 points **A**, **B** et **C**, un **constructeur** qui initialise ce triangle avec 3 points de coordonnées (0, 0), une méthode **perimetre()** qui renvoie la somme des longueurs des segments AB, BC et CA, et une méthode **barycentre()** qui renvoie un Point **G** avec les coordonnées du centre de gravité du triangle ABC.

image.png

```python
class Triangle:
    def __init__(self, A=None, B=None, C=None):
        self.A = A if A else Point()
        self.B = B if B else Point()
        self.C = C if C else Point()

    def perimetre(self):
        AB = Segment(self.A, self.B)
        BC = Segment(self.B, self.C)
        CA = Segment(self.C, self.A)
        return AB.longueur() + BC.longueur() + CA.longueur()

    def barycentre(self):
        G = Point()
        G.x = (self.A.x + self.B.x + self.C.x) / 3
        G.y = (self.A.y + self.B.y + self.C.y) / 3
        return G
```

### B - Tester la classe **Triangle** en créant 3 points **A**, **B** et **C**, puis en déplaçant A vers **(3, 2)**, B vers **(7, 5)**, et C vers **(5, 8)**, puis en affichant le  périmetre du triangle **ABC** et les coordonnées du point **G**, centre de gravité de ce triangle.

```python
A = Point()
B = Point()
C = Point()

A.deplacer(3, 2)
B.deplacer(7, 5)
C.deplacer(5, 8)

T = Triangle(A, B, C)

AB = Segment(A, B)
BC = Segment(B, C)
CA = Segment(C, A)

M_AB = AB.milieu()
M_BC = BC.milieu()
M_CA = CA.milieu()

print("Perimetre =", T.perimetre())

G = T.barycentre()
print("Barycentre =", G.x, G.y)
```
